```python
import tensorflow as tf

def fucao_min():
    return x ** 4 - x ** 3 - x ** 2 + x

x = tf.Variable(1.5)

opt = tf.keras.optimizers.SGD(learning_rate=0.1)
for i in range(100):
    y = fucao_min().numpy()
    print ('{:.0f}° -> y = {:.4f}, x = {:.4f}'.format(i+1 ,y, x.numpy()))
    opt.minimize(fucao_min, var_list=[x])
```
